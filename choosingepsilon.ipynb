```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Laplace PDF function
def laplace_pdf(x, mean, scale_factor):
    return (1 / (2 * scale_factor)) * np.exp(-np.abs(x - mean) / scale_factor)

# Given parameters
mean = 2
epsilon = 1
epsilon1 = 2
# Sensitivities for the two queries
sensitivity_1 = 1
sensitivity_2 = 1

# Compute scale factors based on sensitivity and epsilon
scale_factor_1 = sensitivity_1 / epsilon  # 1 / 1 = 1
scale_factor_2 = sensitivity_2 / epsilon1  # 2 / 1 = 2

# Generate x values for plotting
x_values = np.linspace(-10, 10, 1000)

# Compute the Laplace PDF values for both scale factors
pdf_values_1 = laplace_pdf(x_values, mean, scale_factor_1)
pdf_values_2 = laplace_pdf(x_values, mean, scale_factor_2)

# Plot the PDFs
plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values_1, label='Laplace PDF (sensitivity=1, epsilon=1)', color='blue')
plt.plot(x_values, pdf_values_2, label='Laplace PDF (sensitivity=2, epsilon=1)', color='red', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Laplace Distributions for Different Query Sensitivities')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# Mean and scale factor
mean = 0
scale_factor = 17 / 12

# Generate x values for plotting
x_values = np.linspace(-10, 10, 1000)

# Compute the CDF values for the x values
cdf_values = laplace.cdf(x_values, loc=mean, scale=scale_factor)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(x_values, cdf_values, label='Laplace CDF', color='red')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Laplace Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
